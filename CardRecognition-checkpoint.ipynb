The image dataset given is corrupted and damaged, so I 'll be performing noise removal on them. This is because the model will be predicting from images that it wasn't trained upon, and we are sure to a certain extent that images that will be passed to the model for prediction won't be a corrupted one. So therefore I'll like to train with type of images that will be passed to it for accuracy purpose. From the damage shown, it is clear that the best thing to do the images will be to denoise them. I chose a non linear algorithm which is Median filter because while smoothing the images, it also preserve the edges which is key in this situation.

# Convolutional Neural Network

I will be building a cnn model to classify the data. CNN is apt for an image classification. CNN consist of 3 layers, the input layer, the hidden layer, and the output layer. The hidden layers is what it uses to find pattern and that is the most essential layer of the model. The hidden layer consist of some other layers that allows for feature learning. 
* The convutional layer create a feature map to predict class probability for each feature by applying a filter that scans the whole image. It takes in the filter size and in what stride. It also uses an activation called 'relu'.
* The pooling layer reduced the dimensionality of feature map but retains the most significant information.It can also reduces the no of parameters in a network thereby controlling overfitting.
The ouput layer is a fully connected layer that comprises on the dense a layer which flattens the output generated and determine the class of the image.

Here are some hyper parameters used

* Activation function 'Softmax' - This is because I'm performing a multi-class classification. 

* Loss function 'categorical crossentropy' - Also for the same reason. The loss function measures how well the model is learning. 

* Optimizer 'Adam' - This shows how to reduce the loss and help the model improve guesses.

* I used the 'accuracy' metrics. This metrics shows how well the model is predicting the correct label.

* In the first model, I used Epoch = 50. This specify the no of iteration that should be done.

* This model also have 2 convolutional layers

I implemented a callback function. This helps save the progress of the model.


```python
# I'l  define the model. This model has 2 layers

from tensorflow.python.keras.layers import Input, Dense
from keras import utils as np_utils
from PIL import Image
np.random.seed(100)
import keras
import cv2
import os


INPUT_SHAPE =  (192, 192, 1) # The first 2 argument are the image widht and height. The third argument channel no and its 1 because it a grayscale image
inp = keras.layers.Input(shape=INPUT_SHAPE)

conv1 = keras.layers.Conv2D(32, (3, 3), # The filter here is 32 and a kernel size of 3
                                activation ='relu', padding='same')(inp)
pool1 = keras.layers.MaxPooling2D(pool_size=(2, 2))(conv1)
norm1 = keras.layers.BatchNormalization(axis = -1)(pool1)
drop1 = keras.layers.Dropout(rate=0.2)(norm1)


conv2 = keras.layers.Conv2D(32, (3, 3),
                                activation ='relu', padding='same')(drop1)
pool2 = keras.layers.MaxPooling2D(pool_size=(2, 2))(conv2)
norm2 = keras.layers.BatchNormalization(axis = -1)(pool2)
drop2 = keras.layers.Dropout(rate=0.2)(norm2)

flat = keras.layers.Flatten()(drop2)

hidden1 = keras.layers.Dense(512, activation='relu')(flat)
norm3 = keras.layers.BatchNormalization(axis = -1)(hidden1)
drop3 = keras.layers.Dropout(rate=0.2)(norm3)

hidden2 = keras.layers.Dense(256, activation='relu')(drop3)
norm4 = keras.layers.BatchNormalization(axis = -1)(hidden2)
drop4 = keras.layers.Dropout(rate=0.2)(norm4)

out = keras.layers.Dense(class_no, activation='softmax')(drop4)

model = keras.Model(inputs=inp, outputs=out)
model.compile(optimizer='adam',
               loss='categorical_crossentropy',
               metrics=['accuracy'])
print(model.summary())

```

```python
X_train = X_train/255.0 #I normalized the training set before fitting it into the model to make it more efficient

history = model.fit(X_train,
                    Y_train,
                    batch_size = 10, #This will take the data is batches of 10
                    verbose = 1,   # It will ouput the progress rate
                    epochs=50, 
                    validation_split = 0.2, # This will allow 20% of the data to be used as validation
                    shuffle=False)
print("Test_Accuracy:{:.2f}%".format(model.evaluate(np.array(X_test), np.array(Y_test))[1]*100)) #This evaluate the model using the test set

model.save('cardrecognition_cnn.h5') # This saves the model
```

```python
import itertools  
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
   
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

```python
# Plot the confusion matrix.

cm_plot_labels = [0, 1, 2]
plot_confusion_matrix(cm=cm, classes=cm_plot_labels, title='Confusion Matrix')
```
